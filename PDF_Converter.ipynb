```python
#import camelot
#import camelot.io as camelot
from tabula.io import read_pdf
import pandas as pd
```

```python
tables = read_pdf('/content/drive/MyDrive/Documento_Fernanda.pdf', pages='all')
```

```python
len(tables)
```

```python
tables[1].columns
```

```python
for table in tables:
    print(table.columns)
    print('')
```

```python
df = pd.DataFrame()

for i, table in enumerate(tables[1:]):

    df_temp = table
    print(i)
    print(df_temp.columns)
    print(df_temp.shape)
    if df_temp.shape[0] > 0 and df_temp.shape[1] == 5:
        df_temp.columns = ['NCM', 'DESCRIÇÃO', 'TEC (%)', 'ALÍQUOTA (%)', 'FUNDAMENTAÇÃO']
        df = pd.concat([df, df_temp], axis=0)
```
